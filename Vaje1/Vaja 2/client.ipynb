```python
import socket

#velikost header-ja v bajtih
HEADER_LENGTH = 10

message1 = 'Kratko'.encode('utf-8')
message2 = 'Dooooolgooooo'.encode('utf-8')
```

```python
#padding with spaces up to length = HEADER_LENGTH
header1 =f"{len(message1):<{HEADER_LENGTH}}" #pazi notranji {} potreben, ker je notri spremenljivka!
header2 =f"{len(message2)}"
print(header1, len(header1))
print(header2, len(header2))
```

```python
with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sock:
    #sock.connect(('149.62.71.186', 1235))
    sock.connect(('0.0.0.0', 1235))
    header1 = f"{len(message1):<{HEADER_LENGTH}}".encode('utf-8')
    header2 = f"{len(message2):<{HEADER_LENGTH}}".encode('utf-8')
    sock.sendall(header1+message1+header2+message2)
    message = sock.recv(1024)
    print(message.decode())

    
    
#Sprejmi prvih 10 bajtov, ki ti povejo dolžino sporočila  
#message1length = int(sock.recv(10))
#Preberi sporočilo
#message1 = sock.recv(messsage1length)

#message2length = int(sock.recv(10))
#message2 = sock.recv(messsage2length)
```
